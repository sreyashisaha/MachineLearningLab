```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load the dataset using pandas and display all necessary information contained in the file

```python
# Reading the text file task1.txt and saving it in a dataframe
pd.set_option('display.max_columns', None)
df = pd.read_csv('task1.txt')
```

```python
df.info()
```

Investigate what time of the day it is best to fly so as to have the least possible delays. Does this change with airlines?

```python
# Finding the average delays by adding certain delay columns and storing it in another column
df['AverageDelay'] = df.iloc[:,[15,24,25,26,27,28]].mean(axis=1)
df.head()
```

```python
df['TotalAverageDelays'] = df.iloc[:,[14,15,24,25,26,27,28]].mean(axis=1)
df
```

```python
# Droping rows with nan values for column CRSDepTime
df=df.dropna(subset=['CRSDepTime'])
df
```

```python
# Converting CRSDepTime to String values
df['CRSDepTime'] = df['CRSDepTime'].astype('string')
```

```python
# Counting the length of string for CRSDepTime column
df["DepTimeCount"] = df["CRSDepTime"].str.len()
```

```python
df.info()
```

```python
#df
```

```python
df['CRSDepTime']=df['CRSDepTime'].str.zfill(4)
```

```python
# Slicing and extracting first 2 characters from CRSDeptime and storing it another column Hours
df['Hours'] = df['CRSDepTime'].str.slice(0, 2)
df
```

```python
df2 = df.groupby(['Hours'])['AverageDelay'].min()
# df2
df2.plot(x='Hours',y='AverageDelay')
plt.ylabel("Minimum Average Delay")
plt.title("Delay Vs Hours")
plt.legend()
```

```python
# df2
```

```python
df1 = df.groupby(['Hours', 'UniqueCarrier'],as_index=False)['AverageDelay'].min()
# df5 = df5.to_frame()
# df1.plot()
df1
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.scatterplot(df1.groupby(
[df1['UniqueCarrier'], 'Hours'])['AverageDelay'].mean().unstack(), linewidth=0.5,ax=ax)
plt.title("Minimum Delay WRT AirLines and Time Of The Day")
plt.xlabel('UniqueCarrier')
plt.ylabel('Average Delay')
plt.xticks(rotation=450)
plt.legend()
plt.show()
```

```python
df10 = df.groupby(['Dest','Month','FlightNum'])['FlightNum'].count().reset_index(name="count")
df10
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
Months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.scatterplot(df10.groupby(
[df10['Dest'], 'Month'])['count'].mean().unstack(), linewidth=0.5,ax=ax)
# sns.barplot(TYResult.groupby(
# [TYResult['Dest'], 'Month'])['Min_Delay'].mean().unstack(), linewidth=0.5,ax=ax)
plt.xlabel('Destination')
plt.ylabel('Count Of Flights')
plt.title("Number Of Flights with respect to Destination and Months")
plt.xticks(rotation=450)
plt.legend(Months)
plt.show()
```

As it can be seen from the graph that the number of flights to this destination has been decreased over months and hence it is a negative trend
